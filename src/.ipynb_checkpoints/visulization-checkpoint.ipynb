```python
from feature_extraction import *
from windows import *
import matplotlib.pyplot as plt
```

```python
car_img = cv2.imread('../vehicles/KITTI_extracted/1.png') 
notcar_img = cv2.imread('../non-vehicles/Extras/extra1.png')

car_img = cv2.cvtColor(car_img, cv2.COLOR_BGR2RGB)
notcar_img = cv2.cvtColor(notcar_img, cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.imshow(car_img)
plt.title('Car image')

plt.subplot(1,2,2)
plt.imshow(notcar_img)
plt.title('Not car image')

plt.show()
```

```python
car_img_YCrCb = cv2.cvtColor(car_img, cv2.COLOR_RGB2YCrCb)
car_hog_fea0, car_hog_img0 = get_hog_features(car_img_YCrCb[:,:, 0], 9, 8, 2, True, True)
car_hog_fea1, car_hog_img1 = get_hog_features(car_img_YCrCb[:,:, 0], 9, 8, 2, True, True)
car_hog_fea2, car_hog_img2 = get_hog_features(car_img_YCrCb[:,:, 1], 9, 8, 2, True, True)

plt.figure(figsize = (10,10))
plt.subplot(1,4,1)
plt.imshow(car_img)
plt.title('Car image')

plt.subplot(1,4,2)
plt.imshow(car_hog_img0, cmap = 'gray')
plt.title('Car Y hog image')

plt.subplot(1,4,3)
plt.imshow(car_hog_img1, cmap = 'gray')
plt.title('Car Cr hog image')

plt.subplot(1,4,4)
plt.imshow(car_hog_img2, cmap = 'gray')
plt.title('Car Cb hog image')
plt.show()
```

```python
test_img = cv2.imread('../test_images/test4.jpg')
test_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(test_img)
plt.title('Original test image')
plt.show()
```

```python
image_shape = test_img.shape
print('Image shape:', image_shape)
```

```python
x_start_stop = [0, 1280]
y_start_stop = [380, 500]

windows0 = slide_window(image_shape, x_start_stop, y_start_stop, xy_window=(96, 96), xy_overlap = (0.75, 0.75))
window_image0 = draw_boxes(test_img, windows0)

plt.figure()
plt.imshow(window_image0)
plt.title('Slide window')
plt.show()

x_start_stop = [0, 1280]
y_start_stop = [400, 600]
windows1 = slide_window(image_shape, x_start_stop, y_start_stop, xy_window=(128, 128), xy_overlap = (0.75, 0.75))
window_image1 = draw_boxes(test_img, windows1)
plt.figure()
plt.imshow(window_image1)
plt.title('Slide window')
plt.show()

x_start_stop = [0, 1280]
y_start_stop = [500, 660]
windows2 = slide_window(image_shape, x_start_stop, y_start_stop, xy_window=(160, 160), xy_overlap = (0.75, 0.75))
window_image2 = draw_boxes(test_img, windows2)
plt.figure()
plt.imshow(window_image2)
plt.title('Slide window')
plt.show()

print(len(windows0))
print(len(windows1))
print(len(windows2))
```
